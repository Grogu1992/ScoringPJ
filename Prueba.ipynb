El banco de Bogotá fue fundado en 1870, convirtiéndose en la primera institución financiera del país. En la actualidad funciona como banco comercial de primer piso y hace parte del Grupo aval, cuenta con sedes en 11 países. Como establecimiento de crédito, el Banco de Bogotá tiene como principal objetivo la originación de préstamos mediante la utilización de los recursos de fondeo de los clientes de su pasivo, procurando cumplir con la regulación y normativas vigentes. Para lograr este objetivo, la estructura organizacional del banco se encuentra segmentada en vicepresidencias, que a su vez se componen de gerencias cuyas funciones se centran en cumplir tareas específicas.

Las áreas que intervienen principalmente con el establecimiento de estrategias para la originación de créditos son la Vicepresidencia de Crédito y la Dirección de Riesgo de Crédito y Tesorería. Estas áreas serán las afectadas con el proyecto que se desea realizar con el presente trabajo.


### Área Objetivo
En términos generales, se busca mejorar y acrecentar la colocación de créditos de manera eficiente, es decir disminuyendo el riesgo. El área que implementará las soluciones del presente estudio será la Dirección de Riesgo de Crédito y Tesorería, encargada de la originación masiva de créditos. 
En la actualidad la Dirección es la encargada de establecer las estrategias de aprobación para la cartera masiva del banco, es decir, los clientes más pequeños que son atendidos a través de la Red de Oficinas. 
Los clientes que llegan con una necesidad específica a la Red de Oficinas del banco son atendidos por asesores comerciales. Estos asesores se hacen cargo de perfilar el cliente ofreciéndole el producto financiero que más se ajuste a su necesidad. 
Si el producto pertenece a la gama de productos ofertados en el activo del banco se inicia un proceso de aprobación de crédito. Para el segmento de personas naturales el banco ha venido utilizando desde hace aproximadamente 12 años una herramienta conocida como el “Flujo de Aprobación de Créditos”, a través de la cual, mediante el uso de información del cliente y el uso de algoritmos parametrizados en la herramienta, se generan ofertas relacionadas con la necesidad específica del solicitante facilitando los procesos y tiempos de desembolso.

Los pasos son los siguientes:
1.	El asesor comercial radica una solicitud a nombre del cliente
2.	Se toma información general que utilizan los flujos para hacer una “Pre-aprobación” de acuerdo con el perfil del cliente
3.	Se solicitan los documentos de soporte de la información suministrada por el cliente
4.	Las áreas implicadas realizan la validación documental junto con la corrección de información errónea.
5.  Como proceso intermedio, las solicitudes pasan a un análisis uno a uno llevado a cabo por un área llamada la Dirección de Crédito masivo
6.	Se realiza el proceso de “Aprobación final” utilizando nuevamente como herramienta los flujo
7.	Se realiza el proceso de desembolso

A continuación se muestra el esquema general del flujo de aprobación de créditos: 


Desde hace poco más de 3 años el banco abrió la posibilidad de que el segmento de las personas jurídicas pudieran tener aprobaciones de créditos mediante este flujo. Para esto diseñó una esquema especial alterando el proceso, de manera que se pudiera conservar la metodología de aprobación que se venía siguiendo en los equipos de análisis de crédito para este tipo de clientes, pero que permitiera que la aprobación fuera más rápida. Con la utilización de esta herramienta se planeaba pasar de un tiempo promedio de 30 días en la gestión punta a punta en la generación de un crédito a un promedio de 8 días.
A pesar de que el flujo ha mejorado la gestión de créditos para este tipo de clientes, ciertamente no ha logrado sustituir al canal de aprobación tradicional. Por esta razón el flujo de las personas jurídicas es propenso a tener variadas mejoras. 

Actualmente las aprobaciones en este flujo diseñado para las personas jurídicas dependen casi enteramente del análisis uno a uno que realiza la Dirección de Crédito Masivo, por lo que podría decirse que el criterio de aprobación es netamente subjetivo. 
Para las originaciones que realizan los equipos de crédito tradicional, las medianas y grandes empresas se cuenta con modelos de asignación de scores bastantes robustos que permiten medir numéricamente su riesgo. No obstante, para el grupo de las pequeñas empresas -el grupo objetivo de los flujos de aprobación- la aplicación de estos modelos afecta la asignación de scores de riesgo debido al nivel de detalle de información que se necesita. 

El flujo hoy en día cuenta únicamente con este tipo de scores (llamados los scores de originación) para el segmento de persona natural y utiliza una serie de variables sociodemográficas y de comportamiento con el sistema financiero para esta asignación. En los inicios del flujo de las personas jurídicas, la clasificación anterior no podía ser replicada debido a que este era un nuevo tipo de cliente, por tal motivo se ha vuelto necesario diseñar un modelo de calificación basado en que ya se cuenta con un histórico de información de 3 años 
Entre las ventajas notables del proceso actual es que se está logrando tener una muestra ampliada de perfiles a analizar. Por otro lado, una de las desventajas más sobresalientes es que no se cuenta con una métrica de medición de riesgo.
